Comparing the new Y1H calls done by Anna Berenson, to the previous ones obtained from Juan.

```python
import numpy as np
import pandas as pd

from data_loading import load_y1h_pdi_data
from data_loading import load_valid_isoform_clones

old = load_y1h_pdi_data(add_missing_data=False,
                        include_pY1H_data=False)
old = old.drop_duplicates()
if old['clone_acc'].duplicated().any():
    raise UserWarning('unexpected duplicates')
```

```python
df = pd.read_excel('../data/internal/TF isoforms eY1H calls 14JUL23.xlsx',
              sheet_name='List format')
if df['Bait'].isnull().any():
    raise UserWarning('unexpected missing values')
if df['isoform ID'].isnull().any():
    raise UserWarning('unexpected missing values')
df['Binary'] = df['Binary'].map({'yes': True,
                  'no': False,
                  'inconclusive': np.nan})

y1h = (pd.concat([df.loc[df['Binary'] == True, ["isoform ID"]],
            pd.get_dummies(df["Bait"])],
            axis=1)
        .groupby(["isoform ID"])
        .sum()
        > 0
    ).reset_index()
for _i, row in df.loc[df['Binary'].isnull(), :].iterrows():
    y1h.loc[(y1h['isoform ID'] == row['isoform ID']),
            row['Bait']] = np.nan
    
clones = load_valid_isoform_clones()
if clones['clone_name'].duplicated().any():
    raise UserWarning('unexpected duplicates')
clone_acc = clones.set_index('clone_name')['clone_acc']
gene_symbol = clones.set_index('clone_name')['gene_symbol']
y1h['clone_acc'] = y1h['isoform ID'].map(clone_acc)
y1h['gene_symbol'] = y1h['isoform ID'].map(gene_symbol)
y1h = y1h.drop(columns=['isoform ID'])
y1h = y1h.loc[:, list(y1h.columns[-2:]) + list(y1h.columns[:-2])]
if y1h['clone_acc'].isnull().any():
    raise UserWarning('unexpected missing values')
```

```python
old = old.loc[old['clone_acc'].isin(clones['clone_acc'].values), :]
old_pairs = set()
for _i, row in old.iterrows():
    for bait in old.columns[2:]:
        if row[bait]:
            old_pairs.add((row['clone_acc'], bait))
df['clone_acc'] = df['isoform ID'].map(clone_acc)
rescored_pairs = {(a, b) for a, b in df.loc[:, ['clone_acc', 'Bait']].values}
```

```python
for iso, bait in old_pairs.difference(rescored_pairs):
    if bait not in set(old.columns).difference(set(y1h.columns)):
        print(iso, bait)
```
